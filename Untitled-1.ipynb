# EURUSD Machine Learning Prediction Pipeline

1. **Basic Arithmetic Operations**: Calculating moving averages.
2. **Variable Assignment and Printing**: Storing and displaying the final dataset.
3. **Importing Standard Libraries**: `pandas`, `numpy`, `MetaTrader5` (math equivalent).
4. **Creating a Simple Plot**: Plotting the EURUSD closing prices.

```python
# 3. Importing Standard Libraries
import MetaTrader5 as mt5
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
from datetime import datetime

# 1. Basic Arithmetic Operations & Feature Engineering
def preprocess_data(df):
    data = df.copy()
    
    # Basic arithmetic: Division and Shift for Log Returns
    data['log_return'] = np.log(data['close'] / data['close'].shift(1))
    
    # Moving Averages using basic window operations
    data['SMA_20'] = data['close'].rolling(window=20).mean()
    data['SMA_50'] = data['close'].rolling(window=50).mean()
    
    data['volatility_20'] = data['log_return'].rolling(window=20).std()
    
    # Target Variable: 1 if next period's close is higher, else 0
    data['target'] = (data['close'].shift(-1) > data['close']).astype(int)
    
    data.dropna(inplace=True)
    return data

def connect_and_fetch_data(symbol="EURUSD", timeframe=mt5.TIMEFRAME_H1, num_bars=5000):
    print("Initializing MT5 connection...")
    if not mt5.initialize():
        print(f"Failed to initialize MT5, code: {mt5.last_error()}")
        return None

    if mt5.symbol_info(symbol) is None or not mt5.symbol_select(symbol, True):
        mt5.shutdown()
        return None

    rates = mt5.copy_rates_from_pos(symbol, timeframe, 0, num_bars)
    mt5.shutdown()

    if rates is None: return None
    
    df = pd.DataFrame(rates)
    df['time'] = pd.to_datetime(df['time'], unit='s')
    df.set_index('time', inplace=True)
    
    return df[['open', 'high', 'low', 'close', 'tick_volume']]
```

```python
# 2. Variable Assignment and Printing
# Note: For this to work, ensure the MT5 terminal is open and logged into an account
EURUSD_SYMBOL = "EURUSD"
TIMEFRAME = mt5.TIMEFRAME_M15
HISTORY_BARS = 100000

print(f"Fetching {HISTORY_BARS} bars for {EURUSD_SYMBOL}...")

raw_df = connect_and_fetch_data(symbol=EURUSD_SYMBOL, timeframe=TIMEFRAME, num_bars=HISTORY_BARS)

if raw_df is not None:
    # Process the extracted data
    ml_ready_df = preprocess_data(raw_df)
    
    # Printing results
    print("\n--- Final ML Dataset Sample ---")
    display(ml_ready_df[['close', 'log_return', 'SMA_20', 'target']].tail())
    print(f"\nFinal dataset shape: {ml_ready_df.shape}")
else:
    print("Data extraction failed. Please check your MT5 connection.")
```

```python
# 4. Creating a Simple Plot
if raw_df is not None:
    plt.figure(figsize=(14, 6))
    
    # Plotting the close price and moving averages
    plt.plot(ml_ready_df.index, ml_ready_df['close'], label='EURUSD Close', color='blue', alpha=0.6)
    plt.plot(ml_ready_df.index, ml_ready_df['SMA_50'], label='SMA 50', color='orange', linestyle='--')
    plt.plot(ml_ready_df.index, ml_ready_df['SMA_20'], label='SMA 20', color='green', linestyle='--')
    plt.title('EURUSD Close Price and Moving Averages')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()
```
