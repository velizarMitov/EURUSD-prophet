```python
# 3. Importing Standard Libraries
import MetaTrader5 as mt5
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys

# 1. Basic Arithmetic Operations & Feature Engineering
def preprocess_data(df):
    data = df.copy()
    
    # Basic arithmetic: Division and Shift for Log Returns
    data['log_return'] = np.log(data['close'] / data['close'].shift(1))
    
    # Moving Averages using basic window operations
    data['SMA_20'] = data['close'].rolling(window=20).mean()
    data['SMA_50'] = data['close'].rolling(window=50).mean()
    
    data['volatility_20'] = data['log_return'].rolling(window=20).std()
    
    # Target Variable: 1 if next period's close is higher, else 0
    data['target'] = (data['close'].shift(-1) > data['close']).astype(int)
    
    data.dropna(inplace=True)
    return data

def connect_and_fetch_data(symbol="EURUSD", timeframe=mt5.TIMEFRAME_H1, num_bars=5000):
    from datetime import datetime, timezone
    import time as _time

    print("Initializing MT5 connection...")
    if not mt5.initialize():
        print(f"Failed to initialize MT5. Error code: {mt5.last_error()}")
        print("Ensure the MT5 terminal is open and installed on this machine.")
        return None

    # Check if the symbol exists at all
    symbol_info = mt5.symbol_info(symbol)
    if symbol_info is None:
        print(f"Failed to find symbol '{symbol}'. Error code: {mt5.last_error()}")
        print(f"Look in MT5 Market Watch for the exact spelling (e.g., '{symbol}.pro').")
        mt5.shutdown()
        return None

    # Make symbol visible in Market Watch (required before history download)
    if not symbol_info.visible:
        mt5.symbol_select(symbol, True)
        _time.sleep(1)  # give MT5 a moment to load the symbol

    # Use copy_rates_from (date-based) — more reliable than position-based
    # Fetch backwards from now; MT5 will return up to num_bars available bars
    utc_now = datetime.now(timezone.utc)
    rates = mt5.copy_rates_from(symbol, timeframe, utc_now, num_bars)

    # Fallback: try position-based if date-based fails
    if rates is None or len(rates) == 0:
        print("Date-based fetch returned nothing, trying position-based fallback...")
        rates = mt5.copy_rates_from_pos(symbol, timeframe, 0, num_bars)

    last_err = mt5.last_error()
    mt5.shutdown()

    if rates is None or len(rates) == 0:
        print(f"No rates extracted. MT5 error: {last_err}")
        print("Possible causes:")
        print("  1. Open a EURUSD M15 chart in MT5 to force history download, then retry.")
        print("  2. Go to MT5 -> Tools -> Options -> Charts and set max bars to 2147483647.")
        print("  3. Right-click the chart -> Reload history.")
        return None

    df = pd.DataFrame(rates)
    df['time'] = pd.to_datetime(df['time'], unit='s')
    df.set_index('time', inplace=True)
    df.sort_index(inplace=True)  # ensure chronological order

    return df[['open', 'high', 'low', 'close', 'tick_volume']]
```

```python
# 2. Variable Assignment and Printing
# IMPORTANT: Before running this cell, make sure MT5 Desktop is open!
EURUSD_SYMBOL = "EURUSD"  # Change to "EURUSD.m" or "EURUSD.pro" if your broker uses a suffix
TIMEFRAME = mt5.TIMEFRAME_M15  # M15 timeframe for 15-minute bars

# --- Detect the maximum available bars for M15 automatically ---
mt5.initialize()
bars_available = mt5.terminal_info().maxbars   # terminal's configured max bars limit
# Pull as many bars as possible to count real history depth
_probe = mt5.copy_rates_from_pos(EURUSD_SYMBOL, TIMEFRAME, 0, 200000)
HISTORY_BARS = len(_probe) if _probe is not None else bars_available
mt5.shutdown()

print(f"Terminal max bars setting : {bars_available:,}")
print(f"Actual M15 bars available : {HISTORY_BARS:,}")
print(f"Fetching ALL {HISTORY_BARS:,} M15 bars for {EURUSD_SYMBOL}...")

raw_df = connect_and_fetch_data(symbol=EURUSD_SYMBOL, timeframe=TIMEFRAME, num_bars=HISTORY_BARS)

if raw_df is not None:
    ml_ready_df = preprocess_data(raw_df)

    print("\n--- Final ML Dataset Sample ---")
    display(ml_ready_df[['close', 'log_return', 'SMA_20', 'target']].tail())
    print(f"\nDate range : {ml_ready_df.index[0]}  ->  {ml_ready_df.index[-1]}")
    print(f"Total bars : {len(ml_ready_df):,}")
else:
    print("Data extraction failed. Please review the specific error message above.")
```

# Phase 2 — Model Training & Evaluation

Steps:
1. Load `eurusd_prepared.csv`
2. Split into features `X` and target `y`
3. Scale features with `StandardScaler`
4. Cross-validate with `TimeSeriesSplit` (preserves temporal order)
5. Train a `RandomForestClassifier`
6. Evaluate: accuracy, precision, recall, F1, ROC-AUC
7. Plot feature importances

```python
# Phase 2 — Step 1: Install ML dependencies
import subprocess, sys
subprocess.check_call([sys.executable, "-m", "pip", "install",
                       "scikit-learn", "--quiet"])
print("scikit-learn ready.")
```

```python
# Phase 2 — Step 2: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import (accuracy_score, precision_score,
                             recall_score, f1_score, roc_auc_score,
                             classification_report)

print("All Phase 2 libraries imported.")
```

```python
# Phase 2 — Step 3: Load Dataset
# -----------------------------------------------------------------------
# If you ran Phase 1 in this session, you can also use ml_ready_df directly.
# Otherwise, load from the saved CSV.
# -----------------------------------------------------------------------
CSV_PATH = "eurusd_prepared.csv"   # change path if the file is elsewhere

try:
    df = pd.read_csv(CSV_PATH, index_col=0, parse_dates=True)
    print(f"Loaded from CSV: {df.shape[0]:,} rows, {df.shape[1]} columns")
except FileNotFoundError:
    # Fallback: use the in-memory DataFrame from Phase 1
    if 'ml_ready_df' in dir():
        df = ml_ready_df.copy()
        print(f"CSV not found — using in-memory Phase 1 dataset: {df.shape}")
    else:
        raise FileNotFoundError(
            f"'{CSV_PATH}' not found and no in-memory dataset available.\n"
            "Run Phase 1 first or set CSV_PATH to the correct file path."
        )

print(df.head(3))
print(f"\nTarget distribution:\n{df['target'].value_counts(normalize=True).round(3)}")
```

```python
# Phase 2 — Step 4: Feature / Target Split & Scaling
# -----------------------------------------------------------------------
# Drop OHLCV raw prices — the model should learn from derived features only
# to avoid look-ahead leakage from raw price levels.
# -----------------------------------------------------------------------
FEATURE_COLS = [c for c in df.columns
                if c not in ('target', 'open', 'high', 'low', 'close', 'tick_volume')]

X = df[FEATURE_COLS].values   # shape: (n_samples, n_features)
y = df['target'].values        # shape: (n_samples,)

# StandardScaler: zero mean, unit variance.
# Fit ONLY on training splits inside the CV loop to prevent data leakage.
print(f"Features used ({len(FEATURE_COLS)}): {FEATURE_COLS}")
print(f"X shape: {X.shape}  |  y shape: {y.shape}")
print(f"Class balance — 0: {(y==0).mean():.1%}  |  1: {(y==1).mean():.1%}")
```

```python
# Phase 2 — Step 5: TimeSeriesSplit Cross-Validation + Model Training
# -----------------------------------------------------------------------
# WHY TimeSeriesSplit (not KFold)?
#   Financial time series have temporal dependency: bar N+1 depends on bar N.
#   Standard KFold shuffles data, so a fold can train on FUTURE data and test
#   on PAST data — that is look-ahead bias and will produce wildly optimistic
#   accuracy scores in backtesting.
#
#   TimeSeriesSplit always trains on the past and tests on the future,
#   which mirrors real-world deployment conditions.
#
#   Example with n_splits=5 on 1000 bars:
#     Fold 1: train [0..166]   test [167..333]
#     Fold 2: train [0..333]   test [334..500]
#     ...                      (expanding window)
# -----------------------------------------------------------------------

N_SPLITS  = 5
tscv      = TimeSeriesSplit(n_splits=N_SPLITS)

# Model: RandomForestClassifier
#   - Robust to feature scale differences and outliers
#   - Provides built-in feature importance
#   - n_jobs=-1: use all CPU cores
model = RandomForestClassifier(
    n_estimators=200,
    max_depth=6,          # limit depth to reduce overfitting on noisy price data
    min_samples_leaf=20,  # avoid fitting on tiny leaf nodes
    class_weight='balanced',  # handle class imbalance automatically
    random_state=42,
    n_jobs=-1
)

fold_metrics = []

for fold, (train_idx, test_idx) in enumerate(tscv.split(X), start=1):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    # Fit scaler ONLY on train — transform both
    scaler  = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test  = scaler.transform(X_test)

    # Train
    model.fit(X_train, y_train)

    # Predict
    y_pred  = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    metrics = {
        "fold"     : fold,
        "train_n"  : len(train_idx),
        "test_n"   : len(test_idx),
        "accuracy" : accuracy_score(y_test, y_pred),
        "precision": precision_score(y_test, y_pred, zero_division=0),
        "recall"   : recall_score(y_test, y_pred, zero_division=0),
        "f1"       : f1_score(y_test, y_pred, zero_division=0),
        "roc_auc"  : roc_auc_score(y_test, y_proba),
    }
    fold_metrics.append(metrics)
    print(f"Fold {fold}  |  Acc: {metrics['accuracy']:.3f}  "
          f"F1: {metrics['f1']:.3f}  ROC-AUC: {metrics['roc_auc']:.3f}  "
          f"(train={metrics['train_n']:,} / test={metrics['test_n']:,})")

# Summary across folds
results_df = pd.DataFrame(fold_metrics).set_index("fold")
print("\n--- Cross-Validation Summary ---")
print(results_df[["accuracy","precision","recall","f1","roc_auc"]].round(4))
print("\nMean:")
print(results_df[["accuracy","precision","recall","f1","roc_auc"]].mean().round(4))
```

```python
# Phase 2 — Step 6: Final Model — Train on full dataset, evaluate on last 20%
# -----------------------------------------------------------------------
# After CV confirms the model generalises, retrain on all data except the
# final hold-out period (last 20%) for a clean out-of-sample report.
# -----------------------------------------------------------------------
split_idx = int(len(X) * 0.80)

X_train_full = X[:split_idx]
X_test_full  = X[split_idx:]
y_train_full = y[:split_idx]
y_test_full  = y[split_idx:]

scaler_final  = StandardScaler()
X_train_final = scaler_final.fit_transform(X_train_full)
X_test_final  = scaler_final.transform(X_test_full)

# Recreate model for a clean final fit (independent from CV loop state)
model = RandomForestClassifier(
    n_estimators=200,
    max_depth=6,
    min_samples_leaf=20,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)
model.fit(X_train_final, y_train_full)
y_pred_final  = model.predict(X_test_final)
y_proba_final = model.predict_proba(X_test_final)[:, 1]

print("=== Final Hold-Out Evaluation (last 20% of data) ===")
print(classification_report(y_test_full, y_pred_final,
                             target_names=["Down (0)", "Up (1)"]))
print(f"ROC-AUC : {roc_auc_score(y_test_full, y_proba_final):.4f}")
```

```python
# Phase 2 — Step 7: Feature Importance Plot
importances = pd.Series(model.feature_importances_, index=FEATURE_COLS)
importances = importances.sort_values(ascending=True)

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# --- Left: Feature Importances ---
importances.plot(kind='barh', ax=axes[0], color='steelblue')
axes[0].set_title('Feature Importances (Random Forest)', fontsize=13)
axes[0].set_xlabel('Importance Score')
axes[0].axvline(importances.mean(), color='red', linestyle='--', label='Mean')
axes[0].legend()

# --- Right: CV Metrics per Fold ---
metrics_plot = results_df[["accuracy", "f1", "roc_auc"]]
metrics_plot.plot(kind='bar', ax=axes[1], colormap='tab10', rot=0)
axes[1].set_title('CV Metrics per Fold', fontsize=13)
axes[1].set_xlabel('Fold')
axes[1].set_ylabel('Score')
axes[1].set_ylim(0, 1)
axes[1].axhline(0.5, color='grey', linestyle='--', alpha=0.5, label='Random baseline')
axes[1].legend()

plt.tight_layout()
plt.savefig("eurusd_phase2_results.png", dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved to eurusd_phase2_results.png")
```

# Phase 3 — Advanced Feature Engineering & Non-linear Ensemble Modeling

**Problem:** Phase 2 baseline (RandomForest, 4 features) achieved Accuracy ≈ 0.51 / ROC-AUC ≈ 0.51 — barely better than random chance. Root cause: **high bias** from too few, low-signal features.

**Solution:**
1. **Rich Technical Indicator Suite** — RSI, MACD, Bollinger Bands, ATR, Stochastic, Williams %R, ROC, lagged returns, rolling stats, time-of-day features (20+ features)
2. **Gradient Boosted Trees** — XGBoost & LightGBM are far better at capturing non-linear interactions in noisy financial data than a shallow RandomForest
3. **Side-by-side model comparison** with `TimeSeriesSplit` to objectively select the best model

```python
# Phase 3 — Step 2: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import (accuracy_score, precision_score,
                             recall_score, f1_score, roc_auc_score,
                             classification_report, ConfusionMatrixDisplay)
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

print("All Phase 3 libraries imported.")
```

```python
# Phase 3 — Step 3: Advanced Feature Engineering
# -----------------------------------------------------------------------
# All indicators are computed manually (no extra library dependencies).
# Everything derives solely from past OHLCV data → zero look-ahead bias.
# -----------------------------------------------------------------------

def build_advanced_features(df_raw: pd.DataFrame) -> pd.DataFrame:
    """
    Takes a DataFrame with columns [open, high, low, close, tick_volume]
    and returns an enriched feature set (20+ indicators) plus a binary target.
    """
    df    = df_raw[['open', 'high', 'low', 'close', 'tick_volume']].copy()
    close = df['close']
    high  = df['high']
    low   = df['low']

    # ── Trend / Momentum ──────────────────────────────────────────────

    # Log return: log(P_t / P_{t-1}) — more stationary than raw price
    df['log_return'] = np.log(close / close.shift(1))

    # Simple Moving Averages — trend direction proxies
    df['SMA_20'] = close.rolling(20).mean()
    df['SMA_50'] = close.rolling(50).mean()
    df['SMA_200'] = close.rolling(200).mean()

    # Price relative to each SMA (normalised distance)
    df['dist_SMA_20']  = (close - df['SMA_20'])  / df['SMA_20']
    df['dist_SMA_50']  = (close - df['SMA_50'])  / df['SMA_50']
    df['dist_SMA_200'] = (close - df['SMA_200']) / df['SMA_200']

    # ── Volatility ─────────────────────────────────────────────────────

    # Rolling volatility (realised std of log returns)
    df['volatility_20'] = df['log_return'].rolling(20).std()
    df['volatility_5']  = df['log_return'].rolling(5).std()

    # Average True Range (ATR-14): measure of bar-by-bar volatility
    tr = pd.concat([
        high - low,
        (high - close.shift(1)).abs(),
        (low  - close.shift(1)).abs()
    ], axis=1).max(axis=1)
    df['ATR_14'] = tr.ewm(com=13, adjust=False).mean()

    # Bollinger Bands (20, 2σ)
    bb_mid   = close.rolling(20).mean()
    bb_std   = close.rolling(20).std()
    bb_upper = bb_mid + 2 * bb_std
    bb_lower = bb_mid - 2 * bb_std
    # %B: where is price within the band? 0 = lower band, 1 = upper band
    df['BB_pct_b']     = (close - bb_lower) / (bb_upper - bb_lower)
    # Bandwidth: how wide is the band? High = high volatility regime
    df['BB_bandwidth'] = (bb_upper - bb_lower) / bb_mid

    # ── Oscillators ────────────────────────────────────────────────────

    # RSI(14) — Relative Strength Index (Wilder EMA smoothing)
    delta    = close.diff()
    gain     = delta.clip(lower=0)
    loss     = (-delta).clip(lower=0)
    avg_gain = gain.ewm(com=13, adjust=False).mean()
    avg_loss = loss.ewm(com=13, adjust=False).mean()
    rs       = avg_gain / avg_loss.replace(0, np.nan)
    df['RSI_14'] = 100 - (100 / (1 + rs))

    # MACD (12, 26, 9) — trend-following momentum
    ema12            = close.ewm(span=12, adjust=False).mean()
    ema26            = close.ewm(span=26, adjust=False).mean()
    df['MACD']       = ema12 - ema26
    df['MACD_signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
    df['MACD_hist']  = df['MACD'] - df['MACD_signal']  # histogram = momentum of momentum

    # Stochastic %K and %D (14, 3) — overbought/oversold in recent range
    low14         = low.rolling(14).min()
    high14        = high.rolling(14).max()
    df['Stoch_K'] = 100 * (close - low14) / (high14 - low14)
    df['Stoch_D'] = df['Stoch_K'].rolling(3).mean()      # smoothed %K

    # Williams %R (14) — like Stochastic, inverted scale
    df['Williams_R'] = -100 * (high14 - close) / (high14 - low14)

    # Rate of Change — ROC(10)
    df['ROC_10'] = close.pct_change(10) * 100

    # ── Lagged Returns (micro-memory of recent price path) ─────────────

    for lag in range(1, 6):
        df[f'lag_{lag}'] = df['log_return'].shift(lag)

    # ── Rolling Statistics (short-window context) ──────────────────────

    df['return_mean_5']  = df['log_return'].rolling(5).mean()
    df['return_std_5']   = df['log_return'].rolling(5).std()
    df['return_mean_10'] = df['log_return'].rolling(10).mean()
    df['return_std_10']  = df['log_return'].rolling(10).std()

    # ── Time Features (market session patterns) ────────────────────────
    # FX markets have known intraday patterns (London/NY opens etc.)
    df['hour_of_day'] = df.index.hour
    df['day_of_week'] = df.index.dayofweek   # 0 = Monday, 4 = Friday

    # ── Target ─────────────────────────────────────────────────────────
    # 1 if the NEXT bar closes higher than the current bar, else 0
    df['target'] = (close.shift(-1) > close).astype(int)

    df.dropna(inplace=True)
    return df


# ── Apply to raw OHLCV data ────────────────────────────────────────────
# ml_ready_df (from Phase 1) retains all OHLCV columns, so we use it directly.
# If Phase 1 was not run this session, reload from CSV.
try:
    source_df = ml_ready_df[['open', 'high', 'low', 'close', 'tick_volume']]
except NameError:
    source_df = pd.read_csv("eurusd_prepared.csv", index_col=0, parse_dates=True)

advanced_df = build_advanced_features(source_df)

FEATURE_COLS_P3 = [c for c in advanced_df.columns if c != 'target']

print(f"Advanced dataset : {advanced_df.shape[0]:,} rows × {advanced_df.shape[1]} cols")
print(f"Features ({len(FEATURE_COLS_P3)}): {FEATURE_COLS_P3}")
print(f"\nTarget balance   : {advanced_df['target'].value_counts(normalize=True).round(3).to_dict()}")
advanced_df[FEATURE_COLS_P3[:6]].tail(3)
```

```python
# Phase 3 — Step 4: Multi-Model TimeSeriesSplit Comparison
# -----------------------------------------------------------------------
# WHY these models?
#   - RandomForest    : baseline (same as Phase 2 for fair comparison)
#   - GradientBoosting: sequential boosting — each tree fixes errors of the last
#   - XGBoost         : regularised boosting + histogram split-finding (fast)
#   - LightGBM        : leaf-wise boosting — best for large tabular datasets
#
# Tree-based methods do NOT need feature scaling, so StandardScaler is skipped.
# -----------------------------------------------------------------------

X3 = advanced_df[FEATURE_COLS_P3].values
y3 = advanced_df['target'].values

N_SPLITS = 5
tscv3    = TimeSeriesSplit(n_splits=N_SPLITS)

MODELS = {
    "RandomForest (baseline)": RandomForestClassifier(
        n_estimators=200, max_depth=6, min_samples_leaf=20,
        class_weight='balanced', random_state=42, n_jobs=-1
    ),
    "GradientBoosting": GradientBoostingClassifier(
        n_estimators=300, max_depth=4, learning_rate=0.05,
        subsample=0.8, random_state=42
    ),
    "XGBoost": XGBClassifier(
        n_estimators=300, max_depth=4, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        scale_pos_weight=1,           # class balance handled via scale_pos_weight
        eval_metric='logloss',
        random_state=42, n_jobs=-1, verbosity=0
    ),
    "LightGBM": LGBMClassifier(
        n_estimators=300, max_depth=4, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8,
        class_weight='balanced',
        random_state=42, n_jobs=-1, verbose=-1
    ),
}

all_results = {}   # model_name -> list of fold metric dicts

for model_name, model in MODELS.items():
    fold_metrics = []
    for fold, (train_idx, test_idx) in enumerate(tscv3.split(X3), start=1):
        X_tr, X_te = X3[train_idx], X3[test_idx]
        y_tr, y_te = y3[train_idx], y3[test_idx]

        model.fit(X_tr, y_tr)
        y_pred  = model.predict(X_te)
        y_proba = model.predict_proba(X_te)[:, 1]

        fold_metrics.append({
            "accuracy" : accuracy_score(y_te, y_pred),
            "precision": precision_score(y_te, y_pred, zero_division=0),
            "recall"   : recall_score(y_te, y_pred, zero_division=0),
            "f1"       : f1_score(y_te, y_pred, zero_division=0),
            "roc_auc"  : roc_auc_score(y_te, y_proba),
        })

    all_results[model_name] = pd.DataFrame(fold_metrics)
    mean = all_results[model_name].mean()
    print(f"{model_name:<30}  Acc={mean['accuracy']:.4f}  "
          f"F1={mean['f1']:.4f}  ROC-AUC={mean['roc_auc']:.4f}")

# ── Summary table ──
print("\n=== Mean CV Metrics Across All Folds ===")
summary = pd.DataFrame({
    name: df.mean() for name, df in all_results.items()
}).T.round(4)
print(summary)

# Identify best model by ROC-AUC
BEST_MODEL_NAME = summary['roc_auc'].idxmax()
print(f"\n★ Best model: {BEST_MODEL_NAME}  (ROC-AUC = {summary.loc[BEST_MODEL_NAME,'roc_auc']:.4f})")
```

```python
# Phase 3 — Step 5: Final Hold-Out Evaluation of Best Model
# -----------------------------------------------------------------------
# Retrain the winner on the first 80% of data (chronologically),
# then evaluate on the final unseen 20%.
# -----------------------------------------------------------------------

split_idx = int(len(X3) * 0.80)

X_train_f = X3[:split_idx];  y_train_f = y3[:split_idx]
X_test_f  = X3[split_idx:];  y_test_f  = y3[split_idx:]

# Re-instantiate the best model with the same hyperparameters (clean fit)
best_model = MODELS[BEST_MODEL_NAME]
best_model.fit(X_train_f, y_train_f)

y_pred_f  = best_model.predict(X_test_f)
y_proba_f = best_model.predict_proba(X_test_f)[:, 1]

print(f"=== Final Hold-Out Evaluation: {BEST_MODEL_NAME} ===")
print(f"Hold-out period: {advanced_df.index[split_idx]} → {advanced_df.index[-1]}")
print(f"Test samples   : {len(y_test_f):,}\n")
print(classification_report(y_test_f, y_pred_f, target_names=["Down (0)", "Up (1)"]))
print(f"ROC-AUC : {roc_auc_score(y_test_f, y_proba_f):.4f}")
```

```python
# Phase 3 — Step 6: Visualisations
# -----------------------------------------------------------------------
# Panel A: Model comparison (ROC-AUC per fold for all 4 models)
# Panel B: Feature importance — top 20 features from the best model
# Panel C: Confusion matrix on the final hold-out set
# -----------------------------------------------------------------------

fig = plt.figure(figsize=(20, 14))
gs  = fig.add_gridspec(2, 2, hspace=0.4, wspace=0.35)

# ── Panel A: Per-fold ROC-AUC comparison ──────────────────────────────
ax_a = fig.add_subplot(gs[0, :])   # full width top row

folds     = list(range(1, N_SPLITS + 1))
line_sty  = ['-o', '--s', '-.^', ':D']
for (name, res_df), style in zip(all_results.items(), line_sty):
    ax_a.plot(folds, res_df['roc_auc'], style, label=name, linewidth=2, markersize=7)

ax_a.axhline(0.5, color='grey', linestyle='--', alpha=0.5, label='Random baseline (0.5)')
ax_a.set_title('ROC-AUC per TimeSeriesSplit Fold — All Models', fontsize=14)
ax_a.set_xlabel('Fold (chronological order →)')
ax_a.set_ylabel('ROC-AUC')
ax_a.set_ylim(0.44, None)
ax_a.set_xticks(folds)
ax_a.legend(fontsize=9)
ax_a.grid(True, alpha=0.3)

# ── Panel B: Feature importance (top 20) ──────────────────────────────
ax_b = fig.add_subplot(gs[1, 0])

if hasattr(best_model, 'feature_importances_'):
    importances = pd.Series(best_model.feature_importances_, index=FEATURE_COLS_P3)
    top20 = importances.nlargest(20).sort_values()
    top20.plot(kind='barh', ax=ax_b, color='steelblue')
    ax_b.axvline(top20.mean(), color='red', linestyle='--', linewidth=1, label='Mean')
    ax_b.set_title(f'Top 20 Feature Importances\n({BEST_MODEL_NAME})', fontsize=12)
    ax_b.set_xlabel('Importance Score')
    ax_b.legend(fontsize=8)

# ── Panel C: Confusion matrix ─────────────────────────────────────────
ax_c = fig.add_subplot(gs[1, 1])

ConfusionMatrixDisplay.from_predictions(
    y_test_f, y_pred_f,
    display_labels=["Down (0)", "Up (1)"],
    colorbar=False,
    ax=ax_c,
    cmap='Blues'
)
acc_val = accuracy_score(y_test_f, y_pred_f)
ax_c.set_title(f'Confusion Matrix — Hold-Out Set\n(Accuracy={acc_val:.3f})', fontsize=12)

plt.suptitle(f'Phase 3 Results — {BEST_MODEL_NAME}', fontsize=16, y=1.01, fontweight='bold')
plt.savefig("eurusd_phase3_results.png", dpi=150, bbox_inches='tight')
plt.show()
print("Plot saved to eurusd_phase3_results.png")
```
